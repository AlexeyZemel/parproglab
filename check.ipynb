## Это проверка лабы по параллельному программированию. 

```python
import numpy as np
import matplotlib.pyplot as plt

RESULT_PATH = "result.txt"
MATRIX1_PATH = "matrix1.txt"
MATRIX2_PATH = "matrix2.txt"
STATS_PATH = "stats.txt"


def check():
    matrix1 = np.loadtxt(MATRIX1_PATH, int, skiprows=1)
    matrix2 = np.loadtxt(MATRIX2_PATH, int, skiprows=1)
    res_matrix = np.loadtxt(RESULT_PATH, int, skiprows=1)
    multiply = np.dot(matrix1, matrix2)
    if np.array_equal(multiply, res_matrix):
        print("Matrices are equal")
    else:
        print("Matrices are not equal")
        

def showStats():
    f = open(STATS_PATH).readlines()
    f = [i.split() for i in f]
    x = []
    y = []
    for i in f:
        x.append(int(i[0]))
        y.append(float(i[1]))
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_title("Время от размера (без распараллеливания)")
    plt.plot(x,y, "go-")
    plt.xlabel("Matrix size")
    plt.ylabel("Milliseconds")
    for i,j in zip(x,y):
        # ax.annotate(str(j),xy=(i+10,j+100))
        ax.text(i,j, str(j), ha='center', va='bottom')
    plt.show()


if __name__ == "__main__":
    check()
    showStats()
```
